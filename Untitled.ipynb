```python
skip_n_rows= 3
date = []
value=[]
with open('multiTimeline.csv', encoding='utf8') as f:
    for i,r in enumerate(f):
        if i > skip_n_rows - 1:
            #print(r)
            d,v = r.strip().split(',')
            date.append(d)
            value.append(int(v))
    print(date)
    print('--------------')
    print(value)
```
